```python
data.drop(['dateGame','Home','Unnamed: 0','Away','PTS_H','PTS_A','idGame', 'TEAM_ID_A', 'year', 'TEAM_ID_H'],axis=1,inplace=True)
all_columns = ['Spread','OREB_H','OREB_A','avg_plus_minus_H','avg_plus_minus_A','FG_PCT_A','FG_PCT_H','block_H', 'TO_H', 'heightH', 'weightH', 'ageH', 'block_A', 'TO_A', 'heightA', 'weightA', 'ageA', 'Rivalry.Score', 'FG3_PCT_H', 'FG3_PCT_A', 'Shooting_Advantage', 'Power_Ranking_A', 'Power_Ranking_H', 'dreb_H', 'shots_missed_H', 'dreb_A', 'shots_missed_A', 'Fast_Break_A', 'Fast_Break_H']
data = data[all_columns]
data.columns
```

```python
model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.001), loss='mae')
```

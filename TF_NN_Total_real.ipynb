```python
data = pd.read_csv(r'C:\Users\Barre\Downloads\data_w_everything.csv')

data.drop(['dateGame','Home','Unnamed: 0','Away','PTS_H','PTS_A','idGame', 'TEAM_ID_A', 'year', 'TEAM_ID_H'],axis=1,inplace=True)
all_columns = ['Total','avg_plus_minus_H','avg_plus_minus_A','FG_PCT_A','FG_PCT_H','block_H', 'TO_H', 'heightH', 'weightH', 'ageH', 'block_A', 'TO_A', 'heightA', 'weightA', 'ageA', 'Rivalry.Score', 'FG3_PCT_H', 'FG3_PCT_A', 'Shooting_Advantage', 'Power_Ranking_A', 'Power_Ranking_H', 'dreb_H', 'shots_missed_H', 'dreb_A', 'shots_missed_A', 'Fast_Break_A', 'Fast_Break_H']
data = data[all_columns]
data
```

```python
train_data = data.sample(frac=0.8, random_state=0)
test_data = data.drop(train_data.index)
```

```python
#sns.pairplot(data[['Spread','Unnamed: 0', 'idGame', 'OREB_H', 'block_H', 'TO_H', 'heightH', 'weightH', 'ageH', 'OREB_A', 'block_A', 'TO_A', 'heightA', 'weightA', 'ageA', 'Total', 'OREB', 'Rivalry.Score', 'TEAM_ID_H', 'FG3_PCT_H', 'TEAM_ID_A', 'FG3_PCT_A', 'year', 'Shooting_Advantage', 'Power_Ranking_A', 'Power_Ranking_H', 'dreb_H', 'shots_missed_H', 'dreb_A', 'shots_missed_A', 'Fast_Break_A', 'Fast_Break_H']], diag_kind='kde')
```

```python
train_features = train_data.drop(['Total'], axis=1)
train_labels = train_data['Total']
test_features = test_data.drop(['Total'], axis=1)
test_labels = test_data['Total']
```

```python
normalizer = tf.keras.layers.Normalization()
normalizer.adapt(np.array(train_features))
```

```python
model = tf.keras.Sequential([
    normalizer,
    tf.keras.layers.Dense(1, activation='linear')
])
```

```python
model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.008), loss='mae')
```

```python
model.fit(train_features, train_labels, epochs=400, batch_size=32)
```

```python
test_loss = model.evaluate(test_features, test_labels)
score = model.evaluate(test_features, test_labels, verbose=0)
print(score)
test_loss
```

```python


# Predict the test set
y_pred = model.predict(test_features).flatten() # Flatten the predictions to make them one-dimensional

# Calculate the mean absolute error
mae = np.mean(np.abs(y_pred - test_labels.values)) # Convert test_labels to a numpy array using the .values attribute
print('MAE:', mae)
```
